```python
"""
Creates a corpus from Wikipedia dump file.
Inspired by:
https://github.com/panyang/Wikipedia_Word2vec/blob/master/v1/process_wiki.py
"""

import sys
from gensim.corpora import WikiCorpus

def make_corpus(in_f, out_f):

    """Convert Wikipedia xml dump file to text corpus"""

    output = open(out_f, 'w')
    print('Start!')
    wiki = WikiCorpus(in_f)
    print('Wiki created')

    i = 0
    for text in wiki.get_texts():
        output.write(bytes(' '.join(text), 'utf-8').decode('utf-8') + '\n')
        i = i + 1
        if (i % 10000 == 0):
            print('Processed ' + str(i) + ' articles')
        if (i == 1000000):
            print('Processed ' + str(i) + ' articles')
            break
    output.close()
    print('Processing complete!')
```

```python
question = "Soccer players use their muscle systems to kick a ball into a goal. What organ system coordinates the muscles?"
qid = "ACTAAP_2014_7_6"
```

```python
import wikipedia

ret = wikipedia.search(question, results=4)
print(ret)
for element in ret:
    print(element)
```

```python
from haystack.preprocessor import PreProcessor
from haystack.document_store.elasticsearch import ElasticsearchDocumentStore
processor = PreProcessor(
    clean_empty_lines=True,
    clean_whitespace=True,
    clean_header_footer=True,
    split_by="word",
    split_length=100,
    split_respect_sentence_boundary=True,
    split_overlap=0
)
to_index = []
for element in ret:
    print(element)
    page = wikipedia.page(title=element, pageid=None, auto_suggest=False, redirect=True, preload=False)   
    data = {
                'text': page.content,
                'meta': {}
        }
    docs = processor.process(data)
    to_index = to_index + docs
    # print(docs)
```

```python
document_store = ElasticsearchDocumentStore(host="localhost", username="", password="", index=qid.lower())
```

```python
print(to_index)
```

```python
document_store.get_document_count()
```

```python
document_store.write_documents(to_index,batch_size = 1000)
```

```python
document_store.get_document_count()
```

```python
from haystack.retriever.sparse import ElasticsearchRetriever
retriever = ElasticsearchRetriever(document_store=document_store)
```

```python
for result in retriever.retrieve(query=question, top_k=10):
    print(result.text)
    print("----")
```

```python
from haystack.document_store import FAISSDocumentStore

document_store = FAISSDocumentStore(faiss_index_factory_str="Flat")
```

```python
document_store.write_documents(to_index)
```

```python
from haystack.retriever.dense import DensePassageRetriever
retriever = DensePassageRetriever(document_store=document_store,
                                  query_embedding_model="facebook/dpr-question_encoder-single-nq-base",
                                  passage_embedding_model="facebook/dpr-ctx_encoder-single-nq-base",
                                  max_seq_len_query=64,
                                  max_seq_len_passage=256,
                                  batch_size=16,
                                  use_gpu=True,
                                  embed_title=True,
                                  use_fast_tokenizers=True)
document_store.update_embeddings(retriever)
```

```python
for result in retriever.retrieve(query="What colour is the sun? Yellow", top_k=10):
    print(result.text)
    print("----")
```
